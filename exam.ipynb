```python
#1e

import cv2
import numpy as np
path = r"E:\Images\tajmahal.jpg"
img = cv2.imread(path,0)
cv2.imshow("org",img)
h=img.shape[0]
w=img.shape[1]

for i in range(0,h):
    for j in range(0,w):
        pixel = img[i,j]
        if(pixel>150):
            img[i,j] = np.array([255])
        elif(pixel<50):
            img[i,j] = np.array([0])

cv2.imshow("converted",img)
cv2.waitKey(0)
    
```

```python
#4

import cv2
path = r"E:\Images\tajmahal.jpg"
img = cv2.imread(path)
path2 = r"E:\Images\tajmahal2.jpg"
img2 = cv2.imread(path2)
img2 = cv2.resize(img2,(291,240))
cv2.imshow("org",img)
cv2.imshow("org2",img2)
#arithmetic
add = cv2.add(img,img2)
cv2.imshow("add",add)
sub = cv2.subtract(img,img2)
cv2.imshow("sub",sub)
addw = cv2.addWeighted(img,0.4,img2,0.5,0)
cv2.imshow("addw",addw)
#logical
bitwise_add = cv2.bitwise_and(img,img2,mask=None)
cv2.imshow("bitwise_add",bitwise_add)
bitwise_or = cv2.bitwise_or(img,img2,mask=None)
cv2.imshow("bitwise_or",bitwise_or)
bitwise_not = cv2.bitwise_not(img,img2,mask=None)
cv2.imshow("bitwise_not",bitwise_not)



cv2.waitKey(0)
```

```python
#7 withbackground

import cv2
import numpy as np
path = r"E:\Images\tajmahal.jpg"
img = cv2.imread(path,0)
row,col = img.shape
img2 = np.zeros((row,col),dtype='uint8')
for i in range(row):
    for j in range(col):
        if(img[i,j]>50 and img[i,j]<150):
            img2[i,j]=255
        else:
            img2[i,j]=0
            
cv2.imshow("withbackground",img2)
cv2.waitKey(0)
```

```python
#8

import cv2
from matplotlib import pyplot as plt

image1 = cv2.imread(r"E:\Images\tajmahal.jpg")
image2 = cv2.imread(r"E:\Images\tajmahal.jpg",0)

hist1 = cv2.calcHist(image1,[0],None,[256],[0,256])
hist2 = cv2.calcHist(image1,[1],None,[256],[0,256])
hist3 = cv2.calcHist(image1,[2],None,[256],[0,256])

hist4= cv2.calcHist(image2,[0],None,[256],[0,256])

plt.subplot(221),plt.imshow(image1)
plt.subplot(222),plt.plot(hist1),plt.plot(hist2),plt.plot(hist3)

plt.subplot(223),plt.imshow(image2)
plt.subplot(224),plt.plot(hist4)

plt.show()
```
